# 1) Read the data

```python
import pandas as pd
import numpy as np
df=pd.read_csv("Proteins and samples.csv")
df
```

2) Remove the useless rows

```python
df=df.drop(df.index[0])
new_header = df.iloc[0] #grab the first row for the header
dataframe=df[1:] # take the data less the header row
dataframe.columns = new_header #set the header row as the df header
dataframe
```

3) Transpose the data

```python
dataframe_transposed=dataframe.T
dataframe_transposed
```

4) We want the features or proteins row is the first not the indices

```python
dataframe_transposed.columns = dataframe_transposed.iloc[0] #Set the first row as the header
dataframe_transposed = dataframe_transposed[1:] #Drop the first row since it's now the header
dataframe_transposed.head()
```

5) Put the 'Groups' column which is label in the end of the dataframe

```python
#pop 'Groups' column and store it
groups_column = dataframe_transposed.pop('Groups')
#concatenate the remaining DataFrame with the 'Groups' column at the end
dataframe_transposed = pd.concat([dataframe_transposed, groups_column], axis=1)
dataframe_transposed.head() #to check
```

```python
dataframe_transposed.shape
```

Checking the classes or the groups

```python
print(dataframe_transposed['Groups'].unique())
```

```python
cancerous_samples = dataframe_transposed[dataframe_transposed['Groups']=='Cancerous']
control_samples = dataframe_transposed[dataframe_transposed['Groups']=='Control ']
print(cancerous_samples.shape)
print(control_samples.shape)
```

6) Drop the features with NA value in more than 40% of samples no more than 12 NA.

```python
#define theNA threshold for features (max 40 NA values allowed per feature)
na_threshold=12  #maximum number of NA values allowed per feature
#initialize a list to store features to drop
columns_to_remove=[]
#iterate through all features (excluding 'Groups')
for column in dataframe_transposed.columns:
    if column == 'Groups':
        continue  #skip the 'Groups' column
    #count the number of NAs for this feature
    na_count = dataframe_transposed[column].isna().sum()
    #if the NA count exceeds the threshold, mark the column for removal
    if na_count > na_threshold:
        columns_to_remove.append(column)
print(len(columns_to_remove))
#drop the columns that exceed the NA threshold
filtered_dataframe = dataframe_transposed.drop(columns=columns_to_remove, axis=1)
#display the filtered DataFrame shape
print("Filtered DataFrame Shape:", filtered_dataframe.shape)
```

7) Imputation with the mean of each coloumn for each group

```python
#identify feature columns by excluding the 'Groups' column
feature_columns = filtered_dataframe.columns.difference(['Groups']).tolist()
#proceed with non-group-specific imputation
imputed_dataframe = filtered_dataframe.copy()
#mpute missing values using column-wise mean (non-group-specific)
for column in feature_columns:
    imputed_dataframe[column] = pd.to_numeric(imputed_dataframe[column], errors='coerce')  #ensure numeric
    column_mean = imputed_dataframe[column].mean(skipna=True)  #compute mean ignoring NaNs
    imputed_dataframe[column] = imputed_dataframe[column].fillna(column_mean)  #fill NaNs with the mean
print("Remaining NA values after non-group-specific imputation:")
print(imputed_dataframe.isna().sum().sum())
```

8)Cross-Validation with Feature Selection(MRMR and RFE with comparing with the DEPs from the paper) and Model Evaluation

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve, auc
from mrmr import mrmr_classif
from imblearn.over_sampling import SMOTE
from scipy.stats import zscore
import matplotlib.pyplot as plt
#prepare features and target
X=imputed_dataframe.drop(columns=['Groups'])
y=imputed_dataframe['Groups'].replace({'Cancerous': 1, 'Control ': 0})
#cross-Validation
cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
smote=SMOTE(sampling_strategy='auto', random_state=42)
#initialize results
model_results = {"SVM": [], "Random Forest": []}
roc_auc_results = {"SVM": [], "Random Forest": []}
common_features_all_folds = []
#parameter grids for GridSearchCV
param_grid_svm = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],'gamma': ['scale', 'auto']
}
param_grid_rf = {
    'n_estimators': [50, 100],'max_depth': [None, 10],
    'min_samples_split': [2, 5],'min_samples_leaf': [1, 2]}
#cross-validation loop
for fold, (train_idx, test_idx) in enumerate(cv.split(X, y), 1):
    print(f"Fold {fold}:")
    #split data
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    #Row-Wise Normalization
    normalized_train = X_train.copy()
    for index, row in normalized_train.iterrows():
        row_median = row.median()  #calculate the row's median
        normalized_train.loc[index] =row/row_median  #normalize the row by dividing by its median
    normalized_test = X_test.copy()
    for index, row in normalized_test.iterrows():
        row_median = row.median()  #calculate the row's median
        normalized_test.loc[index] = row / row_median  # Normalize the row by dividing by its median
    #Z-Score Normalization (Column-Wise)
    z_score_train = normalized_train.apply(zscore, axis=0)
    z_score_test = normalized_test.apply(zscore, axis=0)
    #apply SMOTE to the training data
    X_train_resampled, y_train_resampled = smote.fit_resample(z_score_train, y_train)
    #MRMR Feature Selection
    selected_features_mrmr = mrmr_classif(X=pd.DataFrame(X_train_resampled, columns=X.columns), y=y_train_resampled, K=25)
    #RFE Feature Selection
    rfe_estimator = RandomForestClassifier(random_state=42)
    rfe = RFE(estimator=rfe_estimator, n_features_to_select=25, step=1)
    rfe.fit(X_train_resampled, y_train_resampled)
    selected_features_rfe = pd.DataFrame(X_train_resampled, columns=X.columns).columns[rfe.support_]
    #Identify Common Features
    proteins_from_paper = {"P81605", "A0A5C2GDV0", "P05543", "A0A5C2G9R2", "O75882", "A0A0K0K1H8","P43251", "O43866", "P02763", "A0A0A0MS08", "P01824", "Q86YZ3","A0A5C2FTG3", "Q68CN4", "E7ERK6", "D6RBJ7", "A0A5C2GKP8", "P00488",
        "C9JF17", "A0A193CHQ9", "A0A5C2G6C5", "Q08380", "V9HWI6"}
    common_features = set(selected_features_mrmr).intersection(selected_features_rfe).intersection(proteins_from_paper)
    common_features_all_folds.append(common_features)
    print(f"Common Features ({len(common_features)}): {common_features}")
    #common features for training and testing
    X_train_common = pd.DataFrame(X_train_resampled, columns=X.columns)[list(common_features)]
    X_test_common = pd.DataFrame(z_score_test, columns=X.columns)[list(common_features)]
    #Hyperparameter Tuning with GridSearchCV
    #SVM
    grid_search_svm = GridSearchCV(SVC(probability=True, random_state=42), param_grid_svm, scoring='roc_auc', cv=3, n_jobs=-1)
    grid_search_svm.fit(X_train_common, y_train_resampled)
    best_svm = grid_search_svm.best_estimator_
    #Random Forest
    grid_search_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, scoring='roc_auc', cv=3, n_jobs=-1)
    grid_search_rf.fit(X_train_common, y_train_resampled)
    best_rf = grid_search_rf.best_estimator_
    #Evaluate Models
    for model_name, model in {"SVM": best_svm, "Random Forest": best_rf}.items():
        y_pred = model.predict(X_test_common)
        y_proba = model.predict_proba(X_test_common)[:, 1]
        accuracy = accuracy_score(y_test, y_pred)
        roc_auc = roc_auc_score(y_test, y_proba)
        #save results
        model_results[model_name].append(accuracy)
        roc_auc_results[model_name].append(roc_auc)
        print(f"{model_name} - Accuracy: {accuracy:.2f},ROC AUC:{roc_auc:.2f}")
        #plot ROC curve for the fold
        fpr, tpr, _ =roc_curve(y_test, y_proba)
        plt.plot(fpr, tpr, label=f'{model_name} Fold {fold} (AUC = {roc_auc:.2f})')
#final Results
for model_name in model_results:
    print(f"{model_name} Results:")
    print(f"Mean Accuracy: {np.mean(model_results[model_name]):.2f}")
    print(f"Mean ROC AUC: {np.mean(roc_auc_results[model_name]):.2f}")
#plot all ROC curves
plt.plot([0, 1], [0, 1], 'k--', label='Chance')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Across Folds')
plt.legend(loc='lower right')
plt.show()
#common Features Across All Folds
common_features_final = set.intersection(*common_features_all_folds)
print(f"Common Features Across All Folds: {common_features_final}")
```
